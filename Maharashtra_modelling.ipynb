```python
# Removing crops with less number of records

values = ['Other Cereals & Millets','other oilseeds','Mango','Tomato','Total foodgrain','Pulses total','Banana',
          'Onion','Grapes','Tobacco']
df_raw = df_raw.loc[~df_raw['Crop'].isin(values)]
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df_raw['Season'] = le.fit_transform(df_raw['Season'])
```
